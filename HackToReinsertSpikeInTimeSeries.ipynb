```python
from scipy.fft import fft, ifft
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,10,1280)
x = np.cos(t)
y = 500*(np.cos(t+0.4) - 0.995)
y[y<0]=0
x+=y
```

```python
plt.plot(t,x)
```

```python
normaliser = np.convolve(np.ones_like(x),np.blackman(500),mode='same')
plt.plot(x)
np.sum(np.blackman(500))
smoothed = np.convolve(x,np.blackman(500),mode='same')/normaliser
plt.plot(smoothed)
plt.figure()
spike = x - smoothed
plt.plot(spike)
#a = np.argmax(sharp)
#h = np.max(sharp)
#plt.plot(sharp)
```

```python
plt.plot(spike)
loc = np.argmax(spike)
height = np.max(spike)
width =np.std(np.where(spike>0.5)[0])
```

```python
def spikefn(loc,width,height):
    p = np.arange(1280)
    return height * np.exp(-(loc-p)**2/(2*width**2))
#plt.plot(spikefn(loc,width,height))
#plt.plot(spike)
plt.plot(x - spikefn(loc,width,height))
```

```python
res = fft(x - spikefn(loc,width,height))
```

```python
plt.plot(np.linspace(0,1280/100,1280),np.real(res))
plt.xlim([0,2])
plt.grid()
```
