```python
import GPy
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

kernel = GPy.kern.Multidimensional_Integral_Limits(input_dim=4,lengthscale=[0.1,0.1],variances=[1])
X = np.array([[1,0,1,0],[2,1,1,0],[1,0,2,1],[2,1,1.5,1],[2,1,2,1.5]])
y = np.array([[0,3,3,1,3]]).T

m = GPy.models.GPRegression(X,y,kernel)
#m.optimize()
m.plot(plot_limits=[[0,0],[2,2]],fixed_inputs=[(1,0),(3,0)])
```
